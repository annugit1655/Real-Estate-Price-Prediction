# **Exploring the Housing Market in Ames, Iowa**

## **Introduction :**

Over the past few years, housing has changed a lot. Rising incomes, higher living costs, and homeowners wanting better upgrades have reshaped how houses are built and designed. Even rural areas have seen sharp price increases.

Usually when somebody considering to buy house consider the neighborhood, type of house, attributes of the house, type of sale, and time are factored into the price. We will try to understand these factors and how they influence the housing market in Ames, Iowa. We will also try to understand how they may have affected the price of houses over time to see the change that was explained above.

Key Column Categories
1. Identification
Id: Unique property ID.

MSSubClass: Type of dwelling (e.g., 1-story, 2-story, duplex).

2. Location
MSZoning: General zoning classification (Residential, Commercial).

Neighborhood: Physical location within Ames city boundaries.

3. Lot Information
LotFrontage: Linear feet of street connected to property.

LotArea: Lot size in square feet.

Street: Type of road access (Gravel, Paved).

Alley: Type of alley access.

4. House Characteristics
OverallQual: Overall material and finish quality (1–10 scale).

OverallCond: Overall condition rating.

YearBuilt: Original construction date.

YearRemodAdd: Remodel date.

5. Exterior
RoofStyle: Type of roof (Flat, Gable, Hip).

Exterior1st / Exterior2nd: Exterior covering materials.

MasVnrType: Masonry veneer type.

MasVnrArea: Masonry veneer area in square feet.

6. Interior
ExterQual: Exterior quality rating.

BsmtQual: Basement height.

BsmtFinSF1: Finished basement area.

TotalBsmtSF: Total basement area.

HeatingQC: Heating quality and condition.

CentralAir: Central air conditioning (Y/N).

GrLivArea: Above ground living area in square feet.

KitchenQual: Kitchen quality rating.

TotRmsAbvGrd: Total rooms above grade.

7. Garage
GarageType: Garage location (attached, detached).

GarageCars: Size of garage in car capacity.

GarageArea: Garage area in square feet.

8. Other Features
Fireplaces: Number of fireplaces.

PoolArea: Pool area in square feet.

Fence: Fence quality.

MiscFeature: Miscellaneous features (shed, tennis court).

9. Sale Information
SaleType: Type of sale (Warranty deed, contract).

SaleCondition: Condition of sale (Normal, Abnormal).

Target Column: SalePrice.

**This dataset represents the sale of  residential properties in Ames, Iowa, where every single record represents a sale of individual house.
There are identification, location, lot information, house characteristics, exterior, interior, garge, sales information and other features describing each house. The data spans from 2006 to 2010.**
